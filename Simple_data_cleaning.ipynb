```python
# replace 'yes' with 'Yes' for 'self_employed'
X_train_train_dataset.loc[X_train_train_dataset.Self_Employed=='no','Self_Employed'] = 'No'
# replace negative loan amount values with NaN 
X_train_train_dataset.loc[X_train_train_dataset.LoanAmount<0,'LoanAmount'] = np.nan
# replace out of scope Property_Area values with NaN
accepted_property_areas = ['Urban','Rural','Semiurban']
X_train_train_dataset.loc[X_train_train_dataset.Property_Area.map(lambda x: x not in accepted_property_areas),'Property_Area'] = np.nan
```

```python
# handling missing data
from sklearn.impute import SimpleImputer 

X_train_train_dataset_no_nans =  X_train_train_dataset.copy()
# 1. Imputer
number_Imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')  
# 2. Fit the imputer object to the feature matrix (only for numeric features)
number_columns = ['ApplicantIncome', 'CoapplicantIncome',
                'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
number_Imputer = number_Imputer.fit(X_train_train_dataset_no_nans[number_columns])
# 3. Call Transform to replace missing data in X_train_train_dataset (on specific columns) by the mean of the column to which that missing data belongs to
X_train_train_dataset_no_nans[number_columns] = \
  number_Imputer.transform(X_train_train_dataset_no_nans[number_columns]) 
# note column ApplicantIncome in the first row --> before it was a missing value!
X_train_train_dataset_no_nans
```
